# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "/Users/chrisgallegos/Programming/bootcamp/modules/5/PyBer_Analysis/resources/city_data.csv"
ride_data_to_load = "/Users/chrisgallegos/Programming/bootcamp/modules/5/PyBer_Analysis/resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]

#test
total_rides_by_city_type.head()
```

```python
# 2. Get the total drivers for each city type
# The data for this part will come from city_data_df

total_drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]

#test
total_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]

#test
total_fare_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = total_fare_by_city_type / total_rides_by_city_type

#test
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fare_by_city_type / total_drivers_by_city_type

#test
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame()

pyber_summary_df["Total Rides"] = total_rides_by_city_type
pyber_summary_df["Total Drivers"] = total_drivers_by_city_type
pyber_summary_df["Total Fares"] = total_fare_by_city_type
pyber_summary_df["Average Fare per Ride"] = average_fare_per_ride
pyber_summary_df["Average Fare per Driver"] = average_fare_per_driver

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"] = total_fare_by_city_type.map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = average_fare_per_ride.map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = average_fare_per_driver.map("${:,.2f}".format)

#test
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

total_fares_by_date = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]

#test
total_fares_by_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
total_fares_by_date = total_fares_by_date.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_fares_by_date_pivot = total_fares_by_date.pivot(index="date", columns="type", values="fare")

#test
total_fares_by_date_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
total_fares_specific_date = total_fares_by_date_pivot.loc["2019-01-01":"2019-04-29"]

#test
total_fares_specific_date.head()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
total_fares_specific_date.index = pd.to_datetime(total_fares_specific_date.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
total_fares_specific_date.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
total_fares_specific_date = total_fares_specific_date.resample("W").sum()
total_fares_specific_date
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
total_fares_specific_date.plot(figsize=(12,6.6))
plt.ylabel("Fare($USD)")
plt.title("Total Fare by City Type")

# Save the figure.
plt.savefig("/Users/chrisgallegos/Programming/bootcamp/modules/5/PyBer_Analysis/analysis/PyBer_fare_summary.png")
```
